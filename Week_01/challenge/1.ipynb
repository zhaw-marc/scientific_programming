# Header

```python
import numpy as np

arr = np.random.rand(1, 10)
print(arr)
```

```python
quartiles = np.percentile(arr, [25, 50, 75])
print("Quartiles:", quartiles)
```

```python
import matplotlib.pyplot as plt

plt.hist(arr.flatten(), bins=5, edgecolor='black')
plt.title('Histogram of arr')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(arr.flatten())
plt.title('Boxplot of arr')
plt.ylabel('Value')
plt.show()
```
