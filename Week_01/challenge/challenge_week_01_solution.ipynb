# Week 01 Challenge Solution

Exploring random numbers with NumPy: quartiles, histograms, and boxplots.

## Imports & Data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Numpy array with random numbers
np.random.seed(42)
data = np.random.randn(20000) * 15 + 50
print('Array (first 10 values):', data[:10])
print('Shape:', data.shape)
```

## Quartiles

```python
# Quartiles using numpy percentile()
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

print('Q1 (25th percentile):', round(q1, 2))
print('Q2 (50th percentile / Median):', round(q2, 2))
print('Q3 (75th percentile):', round(q3, 2))
print('IQR:', round(q3 - q1, 2))
```

## Histogram

```python
# Histogram
fig, ax = plt.subplots(figsize=(6, 4))
sns.histplot(data, bins=20, color='greenyellow', edgecolor='#1a1a1a', ax=ax)
ax.set_title('Histogram of Random Numbers')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Boxplot

```python
# Boxplot
fig, ax = plt.subplots(figsize=(6, 1.6))
sns.boxplot(
    x=data,
    orient='h',
    color='greenyellow',
    linecolor='auto',
    linewidth=1.5,
    flierprops=dict(marker='o', alpha=0.5),
    ax=ax,
)
ax.set_title('Boxplot of Random Numbers')
ax.set_xlabel('Value')
plt.tight_layout()
plt.show()
```
